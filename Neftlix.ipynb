```python
# show_id--Netflix-in hər kontentə verdiyi unikal ID.
# type---Kontentin növü:"Movie" ,"TV Show"
# title--Filmin və ya serialın adı.
#  director---Filmin və ya serialın rejissoru.
# cast----Kontentdə iştirak edən aktyorlar.Bir neçə aktyor adları vergüllə ayrılıb.
#  country---Filmin və ya serialın istehsal edildiyi ölkə(lər).
# date_added---Kontentin Netflix platformasına əlavə edildiyi tarix.
# release_year----Filmin və ya serialın istehsal olunduğu il.
# rating----Kontentin yaş kateqoriyası:
# duration-----Filmin uzunluğu və ya serialın sezon sayı.
# listed_in-----Kontentin aid olduğu janrlar / kateqoriyalar.
# description-----Filmin və ya serialın qısa xülasəsi.

```

```python
df['date_added']= pd.to_datetime(df['date_added'], format= 'mixed', errors= 'coerce')
```

```python
df['duration_value'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_type'] = df['duration'].str.extract(r'([A-Za-z]+)').astype(str)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['title'] = df['title'].str.strip().str.title()
df['type'] = df['type'].str.strip()
df['country'] = df['country'].str.strip()
```

```python
df['country'] = df['country'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['director'] = df['director'].fillna("Unknown")
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df = df.dropna(subset=['date_added', 'duration'])
```

```python
df.isna().sum()
```

```python
df['type'] = df['type'].astype('category')
df['rating'] = df['rating'].astype('category')
df['duration_type'] = df['duration_type'].astype('category')
```

```python
rating_groups = {
    'Kids': ['TV-Y', 'TV-Y7', 'TV-G', 'G', 'TV-Y7-FV'],
    'Teens': ['TV-PG', 'PG', 'PG-13', 'TV-14'],
    'Adults': ['TV-MA', 'R', 'NC-17', 'NR', 'UR']
}
def cat(x):
  for k ,v in rating_groups.items():
    if x['rating'] in v:
      return k
df=df.copy()
df['category']=df.apply(lambda x : cat(x),axis=1)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type',palette='Set2',legend=False)
plt.title('Movie vs TV Show', fontsize=14)
plt.ylabel('Count')
plt.xlabel('Type')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='category', palette='Set2',hue='type')
plt.title('Content by Age Category', fontsize=14)
plt.ylabel('Count')
plt.xlabel('Category')
plt.show()
```

```python
from textblob import TextBlob
from wordcloud import WordCloud
df['sentiment'] = df['description'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['sentiment_label'] = df['sentiment'].apply(lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral'))

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sentiment_label', palette=['green','red','gray'])
plt.title('Sentiment of Descriptions', fontsize=14)
plt.ylabel('Count')
plt.xlabel('Sentiment')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(df['description'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 2000 və sonrası üçün filter
df_after_2000 = df[df['release_year'] >= 2000]

plt.figure(figsize=(12,5))
sns.countplot(
    data=df_after_2000,
    x='release_year',
    palette='Spectral',
    order=sorted(df_after_2000['release_year'].dropna().unique())
)
plt.xticks(rotation=90)
plt.title('Number of Titles per Year (2000 and After)', fontsize=14)
plt.ylabel('Count')
plt.xlabel('Year')
plt.show()
```
