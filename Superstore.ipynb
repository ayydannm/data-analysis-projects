#Funnel Analysis

```python
#Funnel mÃ¼ÅŸtÉ™rinin alÄ±ÅŸ prosesindÉ™ keÃ§diyi mÉ™rhÉ™lÉ™lÉ™ri gÃ¶stÉ™rir vÉ™ harada itki olduÄŸunu tapmaq Ã¼Ã§Ã¼n istifadÉ™ olunur.
```

```python
# Superstore dataset-i event-level mÉ™lumatlar (visit, add-to-cart, checkout)
# saxlamadÄ±ÄŸÄ± Ã¼Ã§Ã¼n funnel mÉ™rhÉ™lÉ™lÉ™rini mÃ¶vcud sÃ¼tunlardan proxy olaraq yaratmÄ±ÅŸam.
# Bu yanaÅŸma real user davranÄ±ÅŸÄ±nÄ± deyil, biznes mÉ™ntiqinÉ™ É™saslanan tÉ™xmini funnel analizini gÃ¶stÉ™rir.
```

```python
purchase = df[df['Sales'] > 0]['Customer ID'].unique() #satis
checkout = df[df['Profit'] > 0]['Customer ID'].unique() #kassa
cart = df[df['Discount'] > 0]['Customer ID'].unique()  #kart

funnel = {
    'Visited Product Page': len(df['Customer ID'].unique()), #her musteri azi birdefe mehsula baxib
    'Added to Cart': len(set(cart)),
    'Checkout Started': len(set(cart) & set(checkout)),
    'Purchase Completed': len(set(cart) & set(checkout) & set(purchase))
}


# DataFrame ÅŸÉ™klindÉ™
funnel_df = pd.DataFrame.from_dict(funnel, orient='index', columns=['Users'])

# Drop-off faizi
#Drop-off = 1 âˆ’ (Next step users / Current step users)
funnel_df['Drop-off %'] = 1 - funnel_df['Users'].shift(-1) / funnel_df['Users']
funnel_df['Drop-off %'] = funnel_df['Drop-off %'].fillna(0) * 100

print(funnel_df)
```

```python
# ÆgÉ™r Added to Cart â†’ Checkout drop-off bÃ¶yÃ¼kdÃ¼rsÉ™:
# QiymÉ™t problemi
# UX problemi
# ÆgÉ™r Checkout â†’ Purchase drop-off bÃ¶yÃ¼kdÃ¼rsÉ™:
# Ã–dÉ™niÅŸ
# Etibar problemi
```

```python
#MÃ¼ÅŸtÉ™rilÉ™rin zamanla neÃ§É™ faizinin qaldÄ±ÄŸÄ±nÄ± gÃ¶stÉ™rir.
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Ä°lk alÄ±ÅŸ ayÄ±nÄ± tap
df['CohortMonth'] = df.groupby('Customer ID')['Order Date'].transform('min').dt.to_period('M') #ilk 2020.01.01
# SifariÅŸ ayÄ±
df['OrderMonth'] = df['Order Date'].dt.to_period('M') #her sifaris hansi ayda verilib #2020-2,5,8,9,2021-6,7

# Cohort index: neÃ§É™ ay keÃ§ib
df['CohortIndex'] = ((df['OrderMonth'].dt.year - df['CohortMonth'].dt.year) * 12 +
                     (df['OrderMonth'].dt.month - df['CohortMonth'].dt.month) + 1)

# Pivot table: retention matrix
cohort_data = df.groupby(['CohortMonth','CohortIndex'])['Customer ID'].nunique().reset_index()
cohort_counts = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='Customer ID')

cohort_size = cohort_counts.iloc[:,0] #cohort_size hÉ™r cohort-un ilk ayda neÃ§É™ mÃ¼ÅŸtÉ™ri baÅŸladÄ±ÄŸÄ±nÄ± gÃ¶stÉ™rir. SonrakÄ± aylarda qalanlarÄ± bu sayÄ±ya bÃ¶lÃ¼rÃ¼k ki, retention % tapsÄ±n
retention = cohort_counts.divide(cohort_size, axis=0)

print(retention)
```

```python
# df_cohort - sÉ™nin cohort pivot cÉ™dvÉ™lin
plt.figure(figsize=(20,8))
sns.heatmap(retention,
            annot=True,        # hÃ¼ceyrÉ™dÉ™ dÉ™yÉ™ri gÃ¶stÉ™rir
            fmt=".2f",         # 2 onluq rÉ™qÉ™m formatÄ±
            cmap="YlGnBu",     # rÉ™ng sxemi
            cbar=True)

plt.title("Cohort Analysis - Customer Retention", fontsize=18)
plt.xlabel("CohortIndex (Months After First Purchase)")
plt.ylabel("Cohort Month (First Purchase)")
plt.show()
```

#COHORT ANALYSIS

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['order_month'] = df['Order Date'].dt.to_period('M')
```

```python
df['cohort_month'] = df.groupby('Customer ID')['order_month'].transform('min')
```

```python
cohort_data = df.groupby(['cohort_month', 'order_month'])['Customer ID'].nunique().reset_index()

cohort_data['cohort_index'] = (
    cohort_data['order_month'].astype(int) -
    cohort_data['cohort_month'].astype(int)
)
```

```python
cohort_pivot = cohort_data.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='Customer ID'
)

cohort_retention = cohort_pivot.divide(cohort_pivot.iloc[:,0], axis=0)
```

#Churn Analysis

```python
# Churn Analysis â€” mÃ¼ÅŸtÉ™rilÉ™rin (istifadÉ™Ã§ilÉ™rin) xidmÉ™tdÉ™n ayrÄ±lmasÄ±nÄ± analiz etmÉ™kdir. YÉ™ni sual budur:
# â€œKimlÉ™r gedir, niyÉ™ gedir vÉ™ bunu necÉ™ azalda bilÉ™rik?â€
```

```python
last_purchase = (
    df.groupby('Customer ID')['Order Date']
      .max()
      .reset_index()
)

last_date = df['Order Date'].max()

last_purchase['inactive_days'] = (
    last_date - last_purchase['Order Date']
).dt.days
```

```python
last_purchase['churned'] = last_purchase['inactive_days'] > 90
```

```python
churn_rate = last_purchase['churned'].mean() * 100
print(f"Churn Rate: {churn_rate:.1f}%")
```

```python
customer_segment = df[['Customer ID', 'Segment']].drop_duplicates()

churn_seg = last_purchase.merge(customer_segment, on='Customer ID')

churn_by_segment = churn_seg.groupby('Segment')['churned'].mean() * 100
print(churn_by_segment)
```

```python
churn_by_segment.plot(kind='bar', title='Churn Rate by Segment')
plt.ylabel('Churn Rate (%)')
plt.show()
```

```python
customer_region = df[['Customer ID', 'Region']].drop_duplicates()

churn_reg = last_purchase.merge(customer_region, on='Customer ID')

churn_by_region = churn_reg.groupby('Region')['churned'].mean() * 100
churn_by_region.plot(kind='bar', title='Churn Rate by Region')
plt.ylabel('Churn Rate (%)')
plt.show()
```

#RFM

```python
# RFM nÉ™dir? mÃ¼ÅŸtÉ™rilÉ™ri davranÄ±ÅŸÄ±na gÃ¶rÉ™ segmentlÉ™ÅŸdirmÉ™k Ã¼Ã§Ã¼n istifadÉ™ olunan metoddur.

# R (Recency) â€“ mÃ¼ÅŸtÉ™ri nÉ™ qÉ™dÉ™r yaxÄ±n zamanda alÄ±b?

# F (Frequency) â€“ neÃ§É™ dÉ™fÉ™ alÄ±b?

# M (Monetary) â€“ nÉ™ qÉ™dÉ™r pul xÉ™rclÉ™yib?
#Tez-tez, yaxÄ±n zamanda vÉ™ Ã§ox pul xÉ™rclÉ™yÉ™n mÃ¼ÅŸtÉ™ri = É™n dÉ™yÉ™rli mÃ¼ÅŸtÉ™ri
```

```python
snapshot_date = df['Order Date'].max() + pd.Timedelta(days=1)

rfm = df.groupby('Customer ID').agg({
    'Order Date': lambda x: (snapshot_date - x.max()).days,  #Recency
    'Order ID': 'nunique',  #Frequency
    'Sales': 'sum'  #Monetary
}).reset_index()

rfm.columns = ['Customer ID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

```python
rfm['RFM_Segment'] = (
    rfm['R_Score'].astype(str) +
    rfm['F_Score'].astype(str) +
    rfm['M_Score'].astype(str)
)
```

```python
rfm[['R_Score','F_Score','M_Score']] = rfm[['R_Score','F_Score','M_Score']].astype(int)
```

```python
# Champions â†’ É™n dÉ™yÉ™rli mÃ¼ÅŸtÉ™rilÉ™r

# Loyal â†’ aktiv vÉ™ stabil

# At Risk â†’ gedÉ™ bilÉ™r

# Others â†’ normal
```

```python
def segment(row):
    if row['R_Score'] == 5 and row['F_Score'] >= 4:
        return 'Champions'
    elif row['R_Score'] >= 4:
        return 'Loyal'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3:
        return 'At Risk'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment, axis=1)
```

```python
rfm.merge(df[['Customer ID','Sales']], on='Customer ID') \
   .groupby('Segment')['Sales'].sum() \
   .sort_values(ascending=False)
```

#A/B Test

```python
#Others mÃ¼ÅŸtÉ™rilÉ™ri xilas etmÉ™k
```

```python
import numpy as np

others = rfm_churn[rfm_churn['Segment'] == 'Others'].copy()

np.random.seed(42)
others['group'] = np.random.choice(['A', 'B'], size=len(others))
```

```python
np.random.seed(42)

others['converted'] = np.where(
    others['group'] == 'B',
    np.random.binomial(1, 0.30, len(others)),  # B qrupunda 30% conversion ,TÉ™klif alanlar daha Ã§ox alÄ±ÅŸ edir (hipotez)
    np.random.binomial(1, 0.20, len(others))   # A qrupunda 20% conversion
)
```

```python
conversion_rates = others.groupby('group')['converted'].mean()
conversion_rates
```

```python
from statsmodels.stats.proportion import proportions_ztest

success = others.groupby('group')['converted'].sum()
nobs = others.groupby('group')['converted'].count()

z_stat, p_value = proportions_ztest(success, nobs)
p_value
```
